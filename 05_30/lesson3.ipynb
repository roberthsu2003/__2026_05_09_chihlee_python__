```python
# dictionary
students = {
	'chinese':89,
	'english':94,
	'math':65,
	'history':92,
	'discover': 95}
	
print(students)
print(students['chinese'])
print(students['math'])
print(students['discover'])
print(students['english'])
print(students['history'])
```
