```python
from pydantic import BaseModel
class Person(BaseModel):
    first_name: str | None = None
    last_name: str
    age: int = 0

print(Person.model_fields)
```

```python
p = Person(last_name = "Simth")
print(repr(p))
```

```python
class Person(BaseModel):
    first_name: str | None = None
    last_name:str
    age: int = 0
    lucky_number:list[int] = []

print(Person.model_fields)
```

```python
p = Person(last_name='smith', age=0, lucky_number=[1, "2", 3])
print(repr(p))
```

```python
for number in p.lucky_number:
    print(number)
```

```python
from pydantic import Field

data = {
    "id": 100,
    "First_name": "John",
    "LASTNAME": "Smith",
    "age in years": 42
}

class Person(BaseModel):
    id_: int = Field(alias="id")
    first_name: str = Field(alias="First_name")
    last_name: str = Field(alias="LASTNAME")
    age: int = Field(alias="age in years")

p = Person.model_validate(data)
print(repr(p))
```

```python
print(p.model_dump())
```

```python
p.model_dump_json()
```

```python
from datetime import datetime, timezone
from pydantic import Field

class Log(BaseModel):
    dt: datetime = Field(default_factory=lambda: datetime.now(timezone.utc))
    message: str

log1 = Log(message="message 1")
print(repr(log1))
```

```python
log2 = Log(message="message 2")
print(repr(log2))
```

```python
data = {
    "firstName": "Arthur",
    "lastName": "Clarke",
    "born":{
        "place":{
            "country":"Lunar Colony",
            "city": "Central City"
        },
        "date":"2001-01-01"
    }
}
```

```python
from datetime import date
from pydantic import Field, BaseModel

class Place(BaseModel):
    country:str
    city: str

class Born(BaseModel):
    place:Place
    dt:date = Field(alias="date")

class Person(BaseModel):
    first_name:str | None = Field(alias="firstName", default=None)
    last_name:str = Field(alias="lastName")
    born:Born

p = Person.model_validate(data)
print(repr(p))
```

```python
print(p.born.place.city)
```

```python
print(p.model_dump())
```

```python
from typing import List
from pydantic import RootModel

class Pets(RootModel):
    root: List[str]

    def __iter__(self):
        return iter(self.root)

    def __getitem__(self, item):
        return self.root[item]

pets = Pets.model_validate(['dog', 'cat'])
print(pets[0])
print([pet for pet in pets])
```
